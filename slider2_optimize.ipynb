```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import raman as r
from matplotlib.widgets import Slider
degree=10
h=pd.read_csv(r"Raman_plastics\PET_10s_RamanShift__42__536.txt",sep="\t")
h.columns=["x","y"]
x=h["x"].values
x=x[5:]     #include in the module
y1=h["y"].values
y1=y1[5:]
y=r.baselinecorrection(y1,lamda=degree,Zhang=True)     #initial y value

fig,ax = plt.subplots(figsize=(10,5))
plt.subplots_adjust(bottom=0.25)
plt.grid()
#plt.figure(figsize=(15,5))
plt.plot(x,y1,color="red",label="original spectra")
k,=plt.plot(x,y,color="blue",label="base corrected")
plt.legend()
fig.canvas.draw()
fig.canvas.flush_events()
ax_slide=plt.axes([0.25,0.1,0.65,0.03])
deg=Slider(ax_slide, "lamda of Zhangfit", valmin=1,valmax=20,valinit=5,valstep=0.1)

def update(val):
    d=deg.val
    degree=d
    k.set_ydata(r.baselinecorrection(y1,lamda=degree,Zhang=True))
    #k.set_ydata(r.baselinecorrection(y1,degree))
deg.on_changed(update)

#ax.set_xlabel("Raman Shift  (cm^-1)",size="15")
ax.set_ylabel("Intensity   a.u",size="15")

#plt.legend()
plt.show()
```
